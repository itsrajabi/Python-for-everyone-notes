## Why Python?

This chapter provides an overview of Python's role in modern software development, with particular attention to its applications in quantitative finance and data-driven fields. The goal is to understand why Python is a practical and strategic choice for both beginners and professionals.

## 1. Core Characteristics

Python is an interpreted, high-level language known for:
- Readable and concise syntax
- Dynamic typing
- Extensive standard library
- Large ecosystem of third-party packages
- Cross-platform support

These features reduce the time between idea and implementation, making Python suitable for rapid prototyping, scripting, and education.

## 2. Key Application Domains

Python is used across many industries:
- Web development (Django, Flask)
- Data analysis and visualization (pandas, Matplotlib)
- Automation and system administration
- Scientific computing and research
- Education and introductory programming

Two domains where Python has become particularly dominant are **quantitative finance** and **artificial intelligence**. While the latter is widely discussed, the former is a mature, high-stakes field that relies heavily on Python for modeling, risk management, and algorithmic trading.

## 3. Python in Quantitative Finance

Quantitative finance uses mathematical and computational methods to price financial instruments, assess risk, and execute trading strategies. Python is well-suited for this domain because:

- **Rapid experimentation**: Quants can test hypotheses interactively.
- **Numerical libraries**: NumPy, SciPy, and pandas provide efficient data structures.
- **Integration**: Python can call C/C++ code for performance-critical tasks.
- **Visualization**: Libraries like matplotlib and plotly help inspect results.
- **Backtesting frameworks**: Tools like zipline and backtrader simplify strategy validation.

A basic example: computing daily returns from a price series.

```python
# Example: daily returns calculation
prices = [100.0, 102.5, 101.0, 105.0, 107.5]
returns = [(prices[i] - prices[i-1]) / prices[i-1] for i in range(1, len(prices))]
print("Daily returns:", returns)
```

This simple operation is a building block for more advanced models, such as Value at Risk (VaR), option pricing, or pairs trading strategies. Python's ecosystem makes it straightforward to scale such calculations to large datasets and real-time data feeds.

## 4. Python and Artificial Intelligence

Artificial intelligence (AI) and machine learning have also adopted Python as their primary language. Libraries like TensorFlow, PyTorch, and scikit-learn provide production-ready tools for model development and deployment.

While this course does not cover AI in depth, the fundamentals taught here are directly applicable to AI workflows. Data handling, control flow, functions, and object-oriented programming are essential skills for any AI practitioner. In this sense, Python serves as a common foundation for both finance and AI.
