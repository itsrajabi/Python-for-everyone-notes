# Chapter 03: Introducing Lists
## 01 - Changing, Adding, and Removing Elements

In the previous section, we learned how to access individual elements in a list. Now we will explore how to modify lists by changing, adding, and removing elements. These operations are fundamental to working with dynamic data, where the contents of a collection change over time — for example, a shopping cart, a to-do list, or a list of active users.

Lists in Python are **mutable**, meaning you can change their contents after they are created. This flexibility is one of the reasons lists are so widely used.

## 1. Modifying Elements in a List

To change an element, you assign a new value to a specific index.

```python
fruits = ["apple", "banana", "orange"]
fruits[1] = "blueberry"
print(fruits)
```

Output:
```
['apple', 'blueberry', 'orange']
```

You can modify any element as long as the index exists.

```python
scores = [85, 92, 78, 90]
scores[2] = 88  # Update the third score
print(scores)
```

Output:
```
[85, 92, 88, 90]
```

This is useful when you need to update a value, such as correcting a mistake or updating a status.

## 2. Adding Elements to a List

Python provides several ways to add elements to a list. The most common are `append()`, `insert()`, and `extend()`.

### Adding to the End with `append()`

The `append()` method adds a new element to the end of the list.

```python
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
```

Output:
```
['apple', 'banana', 'orange']
```

This is the most common way to build a list dynamically, for example, when collecting user input or processing data.

### Adding at a Specific Position with `insert()`

The `insert()` method adds an element at any index. It takes two arguments: the index and the value to insert.

```python
fruits = ["apple", "banana", "orange"]
fruits.insert(1, "blueberry")  # Insert at position 1
print(fruits)
```

Output:
```
['apple', 'blueberry', 'banana', 'orange']
```

If you insert at index `0`, the new element becomes the first. If you insert at an index beyond the list length, it behaves like `append()`.

### Adding Multiple Elements with `extend()`

The `extend()` method adds all elements from another list (or any iterable) to the end.

```python
fruits = ["apple", "banana"]
more_fruits = ["orange", "grape"]
fruits.extend(more_fruits)
print(fruits)
```

Output:
```
['apple', 'banana', 'orange', 'grape']
```

This is useful for combining two lists.

## 3. Removing Elements from a List

Python provides several ways to remove elements, depending on whether you know the value, the position, or you need to remove the last item.

### Removing by Value with `remove()`

The `remove()` method deletes the first occurrence of a specific value.

```python
fruits = ["apple", "banana", "orange", "banana"]
fruits.remove("banana")  # Removes the first 'banana'
print(fruits)
```

Output:
```
['apple', 'orange', 'banana']
```

If the value does not exist, Python raises a `ValueError`. So it is a good practice to check if the value is present before removing it.

### Removing by Position with `pop()`

The `pop()` method removes the element at a specific index and returns it. If no index is given, it removes and returns the last element.

```python
fruits = ["apple", "banana", "orange"]
removed = fruits.pop(1)  # Removes 'banana'
print(removed)  # banana
print(fruits)   # ['apple', 'orange']
```

```python
last = fruits.pop()  # Removes the last element
print(last)   # orange
print(fruits) # ['apple']
```

`pop()` is useful when you need to process an item and then remove it, like a stack (LIFO).

### Deleting by Position with `del`

The `del` statement removes an element at a specific index or even the entire list. It does not return the removed value.

```python
fruits = ["apple", "banana", "orange"]
del fruits[0]  # Removes 'apple'
print(fruits)  # ['banana', 'orange']
```

You can also delete a slice or the whole list:

```python
del fruits[1:]  # Removes from index 1 to end
del fruits      # Deletes the entire list
```

### Clearing a List with `clear()`

The `clear()` method removes all elements from the list, leaving it empty.

```python
fruits = ["apple", "banana", "orange"]
fruits.clear()
print(fruits)  # []
```

## 4. Practical Examples

Let us see these operations in realistic scenarios.

### Example 1: Managing a To-Do List

```python
tasks = ["Buy groceries", "Clean the house", "Finish report"]

# Add a new task
tasks.append("Call dentist")

# Complete a task (remove it)
completed = tasks.pop(1)  # Removes 'Clean the house'
print(f"Completed: {completed}")

# Insert a high-priority task at the top
tasks.insert(0, "Pay bills")

# Update a task
tasks[2] = "Finish report (urgent)"

print(tasks)
```

Output:
```
Completed: Clean the house
['Pay bills', 'Buy groceries', 'Finish report (urgent)', 'Call dentist']
```

### Example 2: Shopping Cart

```python
cart = ["laptop", "mouse", "keyboard"]

# Customer adds an item
cart.append("monitor")

# Customer changes their mind and removes an item
cart.remove("mouse")

# Customer adds a new item at a specific position (e.g., first)
cart.insert(0, "USB hub")

# Remove the last item (checkout)
last_item = cart.pop()
print(f"Last purchased: {last_item}")

print(cart)
```

Output:
```
Last purchased: monitor
['USB hub', 'laptop', 'keyboard']
```

### Example 3: Processing User Data

```python
usernames = ["alice", "bob", "charlie"]

# A new user registers
usernames.append("diana")

# A user changes their username
usernames[1] = "robert"  # Bob becomes Robert

# A user deletes their account
removed_user = usernames.pop(2)  # Removes charlie
print(f"User '{removed_user}' has left.")

# Add multiple users at once
new_users = ["eve", "frank"]
usernames.extend(new_users)

print(usernames)
```

Output:
```
User 'charlie' has left.
['alice', 'robert', 'diana', 'eve', 'frank']
```

## Summary

- Lists are **mutable**: you can change, add, and remove elements.
- Modify an element by assigning a new value to its index.
- Add elements with `append()` (end), `insert()` (position), or `extend()` (multiple).
- Remove elements with `remove()` (by value), `pop()` (by index, returns value), `del` (by index, no return), or `clear()` (all elements).
- Choose the method based on your need: do you know the value or the position? Do you need the removed item?
- Practice with real-world scenarios like to-do lists, shopping carts, or user management.
