## 02 - Organizing a List

In the previous sections, we learned how to create lists, access elements, and modify their contents. Now we turn to organizing lists — arranging items in a particular order, such as alphabetical, numerical, or reverse.

Organizing data is a fundamental operation in programming. Whether you are sorting a list of names, ranking scores, or preparing data for display, the ability to order elements efficiently is essential. Python provides simple and powerful tools for this purpose.

There are two main approaches to sorting in Python:
- **Permanent sorting**: modifies the original list.
- **Temporary sorting**: returns a new sorted list while leaving the original unchanged.

## 1. Sorting a List Permanently with `sort()`

The `sort()` method arranges the elements of a list in ascending order (alphabetical for strings, numerical for numbers) and modifies the original list permanently. After calling `sort()`, the list is permanently changed.

### Example 1: Sorting a List of Numbers

```python
numbers = [5, 2, 8, 1, 9, 3]
numbers.sort()
print(numbers)
```

Output:
```
[1, 2, 3, 5, 8, 9]
```

### Example 2: Sorting a List of Strings

```python
fruits = ["orange", "apple", "banana", "grape"]
fruits.sort()
print(fruits)
```

Output:
```
['apple', 'banana', 'grape', 'orange']
```

### Example 3: Sorting in Descending Order

You can sort in reverse (descending) order by passing `reverse=True` to `sort()`.

```python
numbers = [5, 2, 8, 1, 9, 3]
numbers.sort(reverse=True)
print(numbers)
```

Output:
```
[9, 8, 5, 3, 2, 1]
```

### Example 4: Sorting a List of Names

```python
names = ["Charlie", "Alice", "Bob", "Diana"]
names.sort()
print(names)
```

Output:
```
['Alice', 'Bob', 'Charlie', 'Diana']
```

### Example 5: Sorting in Reverse Order (Descending Names)

```python
names = ["Charlie", "Alice", "Bob", "Diana"]
names.sort(reverse=True)
print(names)
```

Output:
```
['Diana', 'Charlie', 'Bob', 'Alice']
```

### Important: `sort()` Does Not Return a New List

The `sort()` method modifies the list in place and returns `None`. This is a common source of confusion.

```python
numbers = [3, 1, 4, 2]
result = numbers.sort()
print(result)  # None
print(numbers) # [1, 2, 3, 4]
```

## 2. Sorting a List Temporarily with `sorted()`

The `sorted()` function returns a new sorted list while leaving the original list unchanged. This is useful when you need to preserve the original order for later use.

### Example 1: Sorting Without Changing the Original

```python
numbers = [5, 2, 8, 1, 9, 3]
sorted_numbers = sorted(numbers)
print("Sorted:", sorted_numbers)
print("Original:", numbers)
```

Output:
```
Sorted: [1, 2, 3, 5, 8, 9]
Original: [5, 2, 8, 1, 9, 3]
```

### Example 2: Sorting Strings Temporarily

```python
fruits = ["orange", "apple", "banana", "grape"]
sorted_fruits = sorted(fruits)
print("Sorted:", sorted_fruits)
print("Original:", fruits)
```

Output:
```
Sorted: ['apple', 'banana', 'grape', 'orange']
Original: ['orange', 'apple', 'banana', 'grape']
```

### Example 3: Sorting in Descending Order with `sorted()`

```python
numbers = [5, 2, 8, 1, 9, 3]
descending = sorted(numbers, reverse=True)
print(descending)  # [9, 8, 5, 3, 2, 1]
print(numbers)     # [5, 2, 8, 1, 9, 3] (unchanged)
```

### Example 4: Sorting a List of Mixed Data (All Same Type)

```python
scores = [88, 92, 75, 80, 95]
top_scores = sorted(scores, reverse=True)
print(top_scores)  # [95, 92, 88, 80, 75]
```

## 3. Printing a List in Reverse Order

Sometimes you want to print or process a list in reverse order without permanently sorting it. Python provides the `reverse()` method for this purpose.

### Using `reverse()`

The `reverse()` method reverses the order of the list in place (permanently).

```python
fruits = ["apple", "banana", "orange", "grape"]
fruits.reverse()
print(fruits)
```

Output:
```
['grape', 'orange', 'banana', 'apple']
```

### Using Slicing for a Temporary Reverse

If you want to preserve the original order, you can use slicing to create a reversed copy.

```python
fruits = ["apple", "banana", "orange", "grape"]
reversed_fruits = fruits[::-1]
print("Reversed:", reversed_fruits)
print("Original:", fruits)
```

Output:
```
Reversed: ['grape', 'orange', 'banana', 'apple']
Original: ['apple', 'banana', 'orange', 'grape']
```

### Example: Reversing a List of Numbers

```python
numbers = [10, 20, 30, 40, 50]
numbers.reverse()
print(numbers)  # [50, 40, 30, 20, 10]
```

### Example: Reversing with `reversed()` Function

You can also use the built-in `reversed()` function, which returns an iterator (not a list). You can convert it to a list with `list()`.

```python
fruits = ["apple", "banana", "orange"]
rev = list(reversed(fruits))
print(rev)  # ['orange', 'banana', 'apple']
print(fruits)  # ['apple', 'banana', 'orange'] (unchanged)
```

### Combining Reverse and Sort

A common pattern is to sort in reverse order by using `sort(reverse=True)` or `sorted(reverse=True)`. This is more efficient than sorting and then reversing separately.

```python
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort(reverse=True)
print(numbers)  # [9, 5, 4, 3, 2, 1, 1]
```

## 4. Performance Considerations: Sorting Can Be Slow on Large Lists

Sorting is a computationally intensive operation. The standard sorting algorithm in Python, Timsort, is efficient and runs in **O(n log n)** time on average. This means that as the size of the list grows, the time to sort increases more than linearly.

### Understanding O(n log n)

For a list of size `n`:
- If `n = 100`, sorting takes roughly `100 * log2(100) ≈ 664` operations.
- If `n = 10,000`, sorting takes roughly `10,000 * log2(10,000) ≈ 132,877` operations.
- If `n = 1,000,000`, sorting takes roughly `1,000,000 * log2(1,000,000) ≈ 19,931,568` operations.

This means that sorting a list of one million elements is about **30 times slower** than sorting a list of ten thousand elements, even though the list is only 100 times larger.

### Example: Measuring Sorting Time (Conceptual)

While we won't run this here, you can test it yourself:

```python
import time

# Create a large list
large_list = list(range(1000000))  # 1 million elements

# Time sorting
start = time.time()
large_list.sort()
end = time.time()

print(f"Sorting 1 million elements took {end - start:.2f} seconds")
```

On typical hardware, sorting 1 million integers might take **0.1–0.3 seconds**, while sorting 10 million might take **1–3 seconds**. The time grows quickly.

### When to Be Careful

- **Real-time systems**: If you need to sort data frequently in a real-time application, consider whether you can maintain the list in sorted order or use a data structure designed for quick insertion (like a binary search tree).
- **Large datasets**: When sorting very large datasets (e.g., millions of rows), consider memory usage and time. In some cases, you may need to use more advanced sorting algorithms or external sorting techniques.
- **Unnecessary sorting**: Avoid sorting the same list multiple times. If you need to sort once and then use the sorted result, that is fine. But if you are sorting inside a loop, it can become a performance bottleneck.

### Example: Avoiding Unnecessary Sorting

```python
# Bad: Sorting inside a loop
for i in range(10):
    numbers = [5, 2, 8, 1, 9]
    numbers.sort()  # Does this make sense? Only if numbers change each iteration.
```

In this example, if the list does not change, sorting it repeatedly is wasteful. Sort once outside the loop.

### Example: Sorting a List of Tuples (More Complex)

While we haven't covered tuples yet, sorting complex data can be more expensive. Always consider the data type and the size of the list when deciding to sort.

### Summary of Performance

- `sort()` and `sorted()` use Timsort, which is very efficient for most real-world cases.
- Sorting is generally fast for lists up to tens of thousands of elements.
- For very large lists (millions of elements), sorting can take noticeable time and memory.
- Avoid sorting unnecessarily and consider alternative data structures if sorting is a bottleneck.

## 5. More Examples

Here are several more examples to solidify your understanding.

### Example 1: Sorting Student Scores

```python
scores = [88, 92, 75, 80, 95]
scores.sort()
print("Sorted scores:", scores)
print("Highest:", scores[-1])
print("Lowest:", scores[0])
```

Output:
```
Sorted scores: [75, 80, 88, 92, 95]
Highest: 95
Lowest: 75
```

### Example 2: Sorting Words by Length (Not Covered Yet, But Simple)

```python
words = ["python", "java", "c", "javascript", "go"]
sorted_by_length = sorted(words, key=len)  # key=len is a simple custom sort
print(sorted_by_length)
```

Output:
```
['c', 'go', 'java', 'python', 'javascript']
```

### Example 3: Sorting a List of Dictionaries (Preview)

```python
students = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 22},
    {"name": "Charlie", "age": 28}
]
students.sort(key=lambda x: x["age"])
print(students)
```

Output:
```
[{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}, {'name': 'Charlie', 'age': 28}]
```

### Example 4: Reversing and Then Sorting

```python
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
numbers.reverse()
print(numbers)  # [9, 5, 4, 3, 2, 1, 1]
```

### Example 5: Working with a Shopping Cart

```python
cart_items = ["laptop", "mouse", "keyboard", "monitor"]
cart_items.sort()  # Alphabetical order
print("Sorted cart:", cart_items)

# Check if an item is in the cart (requires sorted list for binary search, but not used here)
print("Is 'mouse' in cart?", "mouse" in cart_items)  # True
```

Output:
```
Sorted cart: ['keyboard', 'laptop', 'monitor', 'mouse']
Is 'mouse' in cart? True
```

### Example 6: Temporarily Sorted for Display

```python
cities = ["Tokyo", "New York", "London", "Paris"]
print("Original:", cities)
print("Sorted (temporary):", sorted(cities))
print("Original still:", cities)
```

Output:
```
Original: ['Tokyo', 'New York', 'London', 'Paris']
Sorted (temporary): ['London', 'New York', 'Paris', 'Tokyo']
Original still: ['Tokyo', 'New York', 'London', 'Paris']
```

## Summary

- Use `sort()` to permanently sort a list in ascending order. Add `reverse=True` for descending.
- Use `sorted()` to create a new sorted list while leaving the original unchanged.
- Use `reverse()` to permanently reverse the order of a list.
- Use slicing `[::-1]` or `reversed()` for a temporary reversed copy.
- Sorting is efficient for most data sizes but can be slow for very large lists (millions of elements).
- Avoid sorting unnecessarily and be mindful of performance in loops or real-time systems.
- Practice with different data types: numbers, strings, and mixed data.
