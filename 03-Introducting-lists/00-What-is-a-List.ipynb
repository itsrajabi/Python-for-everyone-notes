# Chapter 03: Introducing Lists
## 00 - What is a List?

In the previous chapter, we worked with variables that store single values: a number, a string, or a boolean. However, in real-world programming, we often need to work with collections of data. For example, a list of student names, a series of sales figures, or a set of product prices.

In Python, a **list** is a collection of items that are ordered and changeable. It is one of the most versatile and commonly used data structures. Lists allow you to store multiple values in a single variable, access them individually, and perform operations on the entire collection.

## 1. Creating a List

A list is created by placing items inside square brackets `[]`, separated by commas.

```python
# A list of numbers
numbers = [10, 20, 30, 40, 50]

# A list of strings
fruits = ["apple", "banana", "orange"]

# A list with mixed types
mixed = ["Alice", 25, True]
```

Lists can contain any type of data, and even different types within the same list. However, in practice, it is common to keep lists homogeneous (all items of the same type) for clarity and easier processing.

## 2. Accessing Elements in a List

To access an element in a list, you use the list name followed by the index of the element in square brackets.

```python
fruits = ["apple", "banana", "orange"]
first_fruit = fruits[0]  # 'apple'
second_fruit = fruits[1] # 'banana'
print(first_fruit)
print(second_fruit)
```

Output:
```
apple
banana
```

## 3. Index Positions Start at 0, Not 1

One of the most important things to remember about lists is that indexing starts at **0**. The first element is at index `0`, the second at index `1`, and so on. This is a convention shared by many programming languages (including C, Java, and JavaScript) and is known as zero-based indexing.

```python
colors = ["red", "green", "blue", "yellow"]

# Index positions:
# 0: 'red'
# 1: 'green'
# 2: 'blue'
# 3: 'yellow'

print(colors[0])  # red
print(colors[2])  # blue
```

### Why Does Indexing Start at 0?

This convention originates from how memory and pointers work in low-level languages. In C, an array name is a pointer to the first element, and the index represents the offset from that pointer. So index `0` means no offset (the first element). While you do not need to understand memory management in Python, knowing this helps you remember the rule.

### Common Mistake: Index Out of Range

If you try to access an index that does not exist, Python raises an `IndexError`.

```python
fruits = ["apple", "banana"]
print(fruits[2])  # IndexError: list index out of range
```

Always ensure the index you use is within the range of the list length.

## 4. Using Individual Values from a List

Once you have accessed an element, you can use it just like any other variable. You can assign it to a new variable, perform operations on it, or use it in expressions.

```python
scores = [85, 92, 78, 90]
first_score = scores[0]
second_score = scores[1]

average = (first_score + second_score) / 2
print(average)  # 88.5
```

### Using List Elements in Strings

You can combine list elements with strings using concatenation or f-strings.

```python
names = ["Alice", "Bob", "Charlie"]
message = f"The first name is {names[0]}"
print(message)  # The first name is Alice
```

### Accessing from the End (Negative Indexing)

Python also supports negative indexing, which allows you to access elements from the end of the list. The index `-1` refers to the last element, `-2` to the second last, and so on.

```python
fruits = ["apple", "banana", "orange", "grape"]
print(fruits[-1])  # grape (last)
print(fruits[-2])  # orange (second last)
```

This is particularly useful when you do not know the length of the list but need the last element.

### Modifying a Value in a List

You can change an element by assigning a new value to its index.

```python
fruits = ["apple", "banana", "orange"]
fruits[1] = "blueberry"
print(fruits)  # ['apple', 'blueberry', 'orange']
```

## Summary

- A list is a collection of ordered items, created with square brackets `[]`.
- Elements are accessed by their index, starting from `0`.
- Index positions are zero-based, meaning the first element is at index `0`.
- Negative indices allow access from the end (e.g., `-1` for the last).
- Once accessed, list elements can be used like any other variable.
- Always ensure the index exists to avoid `IndexError`.
