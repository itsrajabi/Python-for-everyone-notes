# Chapter 05: If Statements
## 00 - Conditional Tests

In the previous chapters, we learned about variables, lists, and loops. Now we turn to **conditional tests** – the foundation of decision-making in programming.

A **conditional test** is an expression that evaluates to either `True` or `False`. These tests allow programs to make decisions based on certain conditions. Understanding conditional tests is essential before we learn about `if` statements, which use these tests to control the flow of a program.

In this section, we will focus on the different types of conditional tests you can write in Python. The actual `if` statements that use these tests will be covered in the next section.

## 1. Checking for Equality

The most common conditional test checks whether two values are equal. You use the **equality operator** `==` (two equals signs).

### Example 1: Simple Equality

```python
name = "Alice"
print(name == "Alice")  # True
print(name == "Bob")    # False
```

Output:
```
True
False
```

### Example 2: Equality with Numbers

```python
age = 25
print(age == 25)  # True
print(age == 30)  # False
```

Output:
```
True
False
```

### Example 3: Using Variables in Equality Checks

```python
favorite_color = "blue"
user_color = "blue"
print(favorite_color == user_color)  # True

user_color = "red"
print(favorite_color == user_color)  # False
```

Output:
```
True
False
```

## 2. Ignoring Case When Checking for Equality

Python is case-sensitive. This means `"Alice"` and `"alice"` are considered different.

### Example 1: Case Sensitivity

```python
name = "Alice"
print(name == "alice")  # False
print(name == "Alice")  # True
```

Output:
```
False
True
```

### Example 2: Ignoring Case with `.lower()`

To ignore case, convert both strings to the same case using `.lower()` or `.upper()`.

```python
name = "Alice"
print(name.lower() == "alice")  # True
print(name.lower() == "ALICE")  # True
```

Output:
```
True
True
```

### Example 3: User Input with Case Insensitivity

```python
user_input = "YeS"
print(user_input.lower() == "yes")  # True
```

Output:
```
True
```

## 3. Checking for Inequality

The **inequality operator** `!=` checks whether two values are not equal.

### Example 1: Simple Inequality

```python
name = "Alice"
print(name != "Bob")    # True
print(name != "Alice")  # False
```

Output:
```
True
False
```

### Example 2: Inequality with Numbers

```python
age = 25
print(age != 30)  # True
print(age != 25)  # False
```

Output:
```
True
False
```

### Example 3: Practical Use of Inequality

```python
status = "inactive"
print(status != "active")  # True
```

Output:
```
True
```

## 4. Checking Multiple Conditions

Sometimes you need to check more than one condition at the same time. Python provides two logical operators for this: `and` and `or`.

### Using `and`

Both conditions must be `True` for the result to be `True`.

```python
age = 25
name = "Alice"

print(age >= 18 and name == "Alice")  # True (both true)
print(age >= 18 and name == "Bob")    # False (one false)
```

Output:
```
True
False
```

### Using `or`

At least one condition must be `True` for the result to be `True`.

```python
age = 16
has_permission = True

print(age >= 18 or has_permission)  # True (one true)
print(age >= 18 or has_permission)  # True
```

Output:
```
True
True
```

### Example: Combining Multiple Conditions

```python
score = 85
attendance = 90

print(score >= 80 and attendance >= 80)  # True
print(score >= 90 or attendance >= 90)   # True (attendance is true)
print(score >= 90 and attendance >= 90)  # False (score is false)
```

Output:
```
True
True
False
```

## 5. Checking Whether a Value Is in a List

The `in` operator checks if a value exists in a list.

### Example 1: Checking Membership

```python
fruits = ["apple", "banana", "orange"]
print("apple" in fruits)   # True
print("grape" in fruits)   # False
```

Output:
```
True
False
```

### Example 2: Using `in` with Variables

```python
fruits = ["apple", "banana", "orange"]
favorite = "banana"
print(favorite in fruits)  # True
```

Output:
```
True
```

### Example 3: Case Sensitivity with `in`

```python
fruits = ["apple", "banana", "orange"]
print("Apple" in fruits)  # False (case-sensitive)
```

Output:
```
False
```

## 6. Checking Whether a Value Is Not in a List

The `not in` operator checks if a value does not exist in a list.

### Example 1: Checking Non-Membership

```python
fruits = ["apple", "banana", "orange"]
print("grape" not in fruits)  # True
print("apple" not in fruits)  # False
```

Output:
```
True
False
```

### Example 2: Practical Use of `not in`

```python
banned_users = ["alice", "bob", "charlie"]
username = "diana"
print(username not in banned_users)  # True

username = "bob"
print(username not in banned_users)  # False
```

Output:
```
True
False
```

## 7. Boolean Expressions

A **Boolean expression** is any expression that evaluates to either `True` or `False`. The conditional tests we have seen are all Boolean expressions.

### Boolean Values

Python has two Boolean values: `True` and `False`. They are case-sensitive (must be capitalized).

```python
is_active = True
is_deleted = False

print(is_active)    # True
print(is_deleted)   # False
```

### Storing Boolean Results

You can store the result of a conditional test in a variable.

```python
age = 25
is_adult = age >= 18
print(is_adult)  # True
```

Output:
```
True
```

### Example: Boolean Variables in Practice

```python
is_weekend = True
is_holiday = False
can_sleep_in = is_weekend or is_holiday
print(can_sleep_in)  # True

is_weekend = False
can_sleep_in = is_weekend or is_holiday
print(can_sleep_in)  # False
```

Output:
```
True
False
```

## 8. More Examples

Let us combine several conditional tests in practical scenarios.

### Example 1: User Login Validation

```python
username = "john_doe"
password = "secure123"

print(username == "john_doe")  # True
print(password == "secure123")  # True
print(password == "wrongpass")  # False
```

Output:
```
True
True
False
```

### Example 2: Shopping Cart Validation

```python
cart = ["laptop", "mouse", "keyboard"]

print("monitor" in cart)  # False
print("mouse" in cart)    # True
print("tablet" not in cart)  # True
```

Output:
```
False
True
True
```

### Example 3: Age and Permission Check

```python
age = 20
has_license = True

can_drive = age >= 18 and has_license
print(can_drive)  # True
```

Output:
```
True
```

### Example 4: Multiple Conditions with Different Data Types

```python
score = 85
level = "advanced"
is_active = True

print(score >= 80 and level == "advanced")
print(score >= 90 or is_active)
print(level == "beginner" and is_active)
```

Output:
```
True
True
False
```

## Summary

- A **conditional test** is an expression that evaluates to `True` or `False`.
- Use `==` to check for equality and `!=` to check for inequality.
- Python is case-sensitive. Use `.lower()` or `.upper()` to ignore case.
- Use `and` and `or` to check multiple conditions.
- Use `in` to check if a value exists in a list.
- Use `not in` to check if a value does not exist in a list.
- **Boolean expressions** evaluate to `True` or `False`. You can store them in variables.
- Boolean values are `True` and `False` (capitalized).

In the next section, we will learn how to use these conditional tests in `if` statements to control the flow of our programs.
