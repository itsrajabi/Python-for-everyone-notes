# Chapter 05: If Statements
## 01 - If Statements

In the previous section, we learned about conditional tests – expressions that evaluate to `True` or `False`. Now we put those tests to use with **`if` statements**.

An `if` statement allows your program to make decisions. It checks a condition and, if the condition is `True`, executes a block of code. If the condition is `False`, the code is skipped. This is the foundation of logic and decision-making in programming.

The basic structure of an `if` statement is:

```python
if condition:
    # Code to run if condition is True
```

The code block must be indented (4 spaces is the standard).

## 1. Simple `if` Statement

A simple `if` statement runs a block of code only if a condition is `True`.

### Example 1: Checking Age

```python
age = 20

if age >= 18:
    print("You are an adult.")
```

Output:
```
You are an adult.
```

If `age` were 16, nothing would be printed.

### Example 2: Checking a Value

```python
name = "Alice"

if name == "Alice":
    print("Hello, Alice!")
```

Output:
```
Hello, Alice!
```

### Example 3: Using `in` with a List

```python
fruits = ["apple", "banana", "orange"]

if "apple" in fruits:
    print("We have apples!")
```

Output:
```
We have apples!
```

## 2. `if-else` Statement

An `if-else` statement provides an alternative block of code to run when the condition is `False`.

```python
if condition:
    # Code to run if condition is True
else:
    # Code to run if condition is False
```

### Example 1: Age Check with Else

```python
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

Output:
```
You are a minor.
```

### Example 2: Checking Membership

```python
fruits = ["apple", "banana", "orange"]

if "grape" in fruits:
    print("We have grapes!")
else:
    print("We do not have grapes.")
```

Output:
```
We do not have grapes.
```

### Example 3: Even or Odd

```python
number = 7

if number % 2 == 0:
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")
```

Output:
```
7 is odd.
```

## 3. The `if-elif-else` Chain

When you have more than two possible outcomes, you can use the `if-elif-else` chain. `elif` is short for "else if".

```python
if condition1:
    # Code for condition1
elif condition2:
    # Code for condition2
else:
    # Code if none of the conditions are true
```

Python evaluates the conditions in order. The first condition that is `True` runs its block, and the rest are skipped.

### Example 1: Grading System

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade is: {grade}")
```

Output:
```
Your grade is: B
```

### Example 2: Ticket Pricing

```python
age = 25

if age < 12:
    price = 5
elif age < 18:
    price = 8
elif age < 65:
    price = 12
else:
    price = 7

print(f"Ticket price: ${price}")
```

Output:
```
Ticket price: $12
```

### Example 3: Weather Advice

```python
temperature = 15

if temperature > 30:
    advice = "It is hot. Stay hydrated."
elif temperature > 20:
    advice = "It is warm. Enjoy the weather."
elif temperature > 10:
    advice = "It is cool. Bring a jacket."
else:
    advice = "It is cold. Bundle up."

print(advice)
```

Output:
```
It is cool. Bring a jacket.
```

## 4. Using Multiple Conditions

You can combine conditions using `and` and `or` inside an `if` statement.

### Example 1: Using `and`

```python
age = 22
has_id = True

if age >= 18 and has_id:
    print("You can enter the venue.")
else:
    print("You cannot enter.")
```

Output:
```
You can enter the venue.
```

### Example 2: Using `or`

```python
is_weekend = True
is_holiday = False

if is_weekend or is_holiday:
    print("You can sleep in.")
else:
    print("Time to wake up for work.")
```

Output:
```
You can sleep in.
```

### Example 3: Multiple Conditions with `in`

```python
fruits = ["apple", "banana", "orange"]

if "apple" in fruits and "banana" in fruits:
    print("We have both apples and bananas.")
```

Output:
```
We have both apples and bananas.
```

## 5. Testing Multiple Conditions

Sometimes you need to test multiple independent conditions, and you want each one to be checked separately (not in an `if-elif-else` chain where only one runs). In this case, use separate `if` statements.

### Example 1: Multiple Independent Checks

```python
toppings = ["pepperoni", "mushrooms", "olives"]

if "pepperoni" in toppings:
    print("Added pepperoni.")
if "mushrooms" in toppings:
    print("Added mushrooms.")
if "olives" in toppings:
    print("Added olives.")
if "onions" in toppings:
    print("Added onions.")
```

Output:
```
Added pepperoni.
Added mushrooms.
Added olives.
```

Notice that all three conditions that are `True` run their code. This is different from `if-elif-else`, where only the first `True` condition runs.

### Example 2: User Preferences

```python
user_settings = {
    "notifications": True,
    "dark_mode": True,
    "auto_save": False
}

if user_settings["notifications"]:
    print("Notifications are enabled.")
if user_settings["dark_mode"]:
    print("Dark mode is enabled.")
if user_settings["auto_save"]:
    print("Auto-save is enabled.")
```

Output:
```
Notifications are enabled.
Dark mode is enabled.
```

### Example 3: Checking Multiple Values in a List

```python
available_items = ["laptop", "mouse", "keyboard", "monitor"]

if "laptop" in available_items:
    print("Laptop is in stock.")
if "mouse" in available_items:
    print("Mouse is in stock.")
if "tablet" in available_items:
    print("Tablet is in stock.")
```

Output:
```
Laptop is in stock.
Mouse is in stock.
```

## 6. Comprehensive Example

Let us combine everything we have learned so far – variables, lists, loops, and `if` statements – into a practical example.

### Scenario: Student Grade Management

We have a list of student names and their scores. We need to:
1. Calculate a letter grade for each student.
2. Identify students who passed (score >= 70).
3. Print a summary report.
4. Check if any student scored 100 (perfect score).
5. Check if any student failed (score < 60) and list them separately.

```python
# Data
students = [
    {"name": "Alice", "score": 92},
    {"name": "Bob", "score": 78},
    {"name": "Charlie", "score": 65},
    {"name": "Diana", "score": 88},
    {"name": "Eve", "score": 55},
    {"name": "Frank", "score": 100}
]

passed = []
failed = []
perfect_scores = []

print("=== Grade Report ===\n")

for student in students:
    name = student["name"]
    score = student["score"]
    
    # Calculate grade using if-elif-else
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    
    print(f"{name}: {score} -> Grade {grade}")
    
    # Track passed/failed using if-else
    if score >= 70:
        passed.append(name)
    else:
        failed.append(name)
    
    # Check for perfect score using simple if
    if score == 100:
        perfect_scores.append(name)

# Summary section with multiple conditions
print("\n=== Summary ===")

# Check passed/failed using if-elif-else
if passed:
    print(f"✅ Students who passed ({len(passed)}): {', '.join(passed)}")
else:
    print("No students passed.")

if failed:
    print(f"❌ Students who failed ({len(failed)}): {', '.join(failed)}")
else:
    print("No students failed.")

# Multiple independent checks
if perfect_scores:
    print(f"🎉 Perfect scores: {', '.join(perfect_scores)}")
else:
    print("No perfect scores.")

# Check if everyone passed
if len(passed) == len(students):
    print("🎊 Everyone passed!")
elif len(failed) == len(students):
    print("😢 Everyone failed.")
else:
    print("📊 Some students passed and some failed.")
```

Output:
```
=== Grade Report ===

Alice: 92 -> Grade A
Bob: 78 -> Grade C
Charlie: 65 -> Grade D
Diana: 88 -> Grade B
Eve: 55 -> Grade F
Frank: 100 -> Grade A

=== Summary ===
✅ Students who passed (4): Alice, Bob, Diana, Frank
❌ Students who failed (2): Charlie, Eve
🎉 Perfect scores: Frank
📊 Some students passed and some failed.
```

## Summary

- `if` statements allow your program to make decisions based on conditions.
- A simple `if` runs code when a condition is `True`.
- `if-else` provides an alternative when the condition is `False`.
- `if-elif-else` handles multiple possible outcomes (only the first `True` runs).
- Use `and` and `or` to combine multiple conditions in one `if` statement.
- Use separate `if` statements to test independent conditions (multiple can run).
- Indentation is critical. Use 4 spaces consistently.
- You can combine `if` statements with loops, lists, and variables to build powerful programs.
