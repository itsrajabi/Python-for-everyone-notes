## 01 - Numbers

In the previous section, we worked with strings and learned how to store and manipulate text. Now we turn to numbers, which are equally essential in programming. Numbers are used for calculations, counting, measurements, and much more. Understanding how Python handles numbers is fundamental to writing useful programs.

Python supports several numeric types, but the two most common are **integers** and **floats**. Before diving into them, let us briefly discuss data types in general.

## 1. Data Types and the `type()` Function

Every value in Python has a data type. The data type determines what operations can be performed on the value. You can check the type of any variable using the built-in `type()` function.

```python
name = "Alice"
age = 30
height = 5.8
is_student = True

print(type(name))      # <class 'str'>
print(type(age))       # <class 'int'>
print(type(height))    # <class 'float'>
print(type(is_student))# <class 'bool'>
```

Knowing the type of a variable is useful for debugging and understanding why certain operations work or fail. It also helps you write more robust code by being aware of what data you are dealing with.

## 2. Integers (`int`)

Integers are whole numbers without a decimal point. They can be positive, negative, or zero.

```python
x = 10
y = -3
z = 0

print(x + y)   # 7
print(x * y)   # -30
print(x ** 2)  # 100 (exponentiation)
```

Integers support all standard arithmetic operations: addition (`+`), subtraction (`-`), multiplication (`*`), division (`/`), floor division (`//`), modulus (`%`), and exponentiation (`**`).

### Division of Integers

In Python 3, division (`/`) always returns a float, even if the result is a whole number.

```python
print(10 / 2)   # 5.0 (float)
print(10 / 3)   # 3.3333333333333335 (float)
```

If you want integer division (floor division), use `//`.

```python
print(10 // 3)  # 3 (int)
print(10 // 2)  # 5 (int)
```

Floor division discards the fractional part and returns the largest integer less than or equal to the result. This is useful when you need a whole number, such as when counting items or indexing.

## 3. Floats (`float`)

Floats are numbers with a decimal point. They are used to represent real numbers.

```python
pi = 3.14159
temperature = -2.5
speed = 0.001

print(pi + 1)       # 4.14159
print(pi * 2)       # 6.28318
print(temperature / 2)  # -1.25
```

Floats are essential for scientific calculations, financial applications, and any domain that requires precision beyond whole numbers.

### The Problem with Floating-Point Arithmetic

One common issue with floats is that they are not always perfectly accurate. This is due to how computers represent decimal numbers in binary. For example:

```python
print(0.1 + 0.2)
```

Output:
```
0.30000000000000004
```

Instead of `0.3`, you get `0.30000000000000004`. This is a small rounding error, but it can accumulate in calculations and lead to unexpected results. This is not a bug in Python; it is a limitation of floating-point arithmetic that exists in most programming languages.

### How to Handle Floating-Point Errors

For most applications, this small error is acceptable. However, if you need exact decimal calculations (e.g., financial calculations), consider using the `decimal` module.

```python
from decimal import Decimal
print(Decimal('0.1') + Decimal('0.2'))  # 0.3
```

For now, be aware that floats are approximate. When comparing floats, avoid checking for exact equality. Instead, check if the difference is within a small tolerance.

```python
a = 0.1 + 0.2
b = 0.3
print(abs(a - b) < 1e-9)  # True
```

## 4. Avoiding Type Errors with the `str()` Function

Type errors occur when you try to perform an operation on incompatible data types. A common example is trying to concatenate a string with a number using the `+` operator.

```python
age = 25
message = "I am " + age + " years old."
```

This raises a `TypeError` because Python does not know how to combine a string (`"I am "`) with an integer (`age`). To fix this, you must explicitly convert the number to a string using the `str()` function.

```python
age = 25
message = "I am " + str(age) + " years old."
print(message)
```

Output:
```
I am 25 years old.
```

### Common Mistakes with `str()`

1. **Forgetting to convert**: This is the most common mistake. Beginners often forget to wrap numeric variables with `str()` when building strings.

```python
# Mistake
total = 150
print("Total: " + total)  # TypeError

# Correct
print("Total: " + str(total))
```

2. **Using `str()` unnecessarily**: Sometimes beginners use `str()` on values that are already strings, which is unnecessary.

```python
name = "Alice"
print(str(name))  # Works but redundant
```

3. **Concatenating without converting all numbers**: If you have multiple numeric values, each must be converted individually.

```python
a = 10
b = 20
print("Sum: " + str(a) + str(b))  # Incorrect: "Sum: 1020"
print("Sum: " + str(a + b))       # Correct: "Sum: 30"
```

4. **Using `str()` with `print()`**: `print()` accepts multiple arguments and automatically converts them to strings. You do not need `str()` inside `print()` for simple cases.

```python
age = 25
print("I am", age, "years old.")  # Works without str()
```

### Using f-strings (Recommended)

The cleanest way to avoid type errors is to use f-strings. They handle type conversion automatically and are more readable.

```python
age = 25
message = f"I am {age} years old."
print(message)
```

f-strings are the preferred way to build strings with embedded values in modern Python.

## 5. Integers in Python 2 (Historical Note)

Python 2 is outdated and no longer supported, but you may encounter it in legacy code. In Python 2, division of two integers performed floor division by default, returning an integer.

```python
# Python 2
print(10 / 3)  # 3 (int), not 3.333...
```

In Python 3, division always returns a float. If you need integer division in Python 3, use `//`. This change was made to make Python more intuitive and consistent.

## Summary

- Python has multiple data types. Use `type()` to check a variable's type.
- **Integers** are whole numbers and support arithmetic operations.
- **Floats** are decimal numbers but have small rounding errors due to binary representation.
- Use `str()` to convert numbers to strings when concatenating.
- f-strings provide a cleaner way to embed values in strings and avoid type errors.
- Python 2 handled integer division differently, but Python 3 uses floats for division.
