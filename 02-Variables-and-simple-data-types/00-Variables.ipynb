## 00 - Variables

In mathematics, a variable is a symbol that represents a value. The equation `x = 5` assigns the value `5` to the symbol `x`. In programming, variables work the same way. They store values in memory so that you can refer to them, manipulate them, and reuse them throughout your program.

Variables are the foundation of programming because they allow you to write flexible, dynamic code. Without variables, every value would be hard-coded, and programs would be static and repetitive. With variables, you can write code that adapts to different inputs, processes data, and performs calculations — which is essentially what programming is about.

Python is a dynamically typed language. This means you do not need to declare the type of a variable before using it. Python infers the type from the value you assign.

## 1. Naming and Using Variables

Variable names in Python must follow a few simple rules:
- They can contain letters, digits, and underscores (`_`).
- They must start with a letter or an underscore, not a digit.
- They are case-sensitive (`age` and `Age` are different).
- They cannot be Python reserved keywords (`if`, `for`, `while`, `class`, etc.).

Beyond the rules, there are conventions that make code more readable:
- Use lowercase letters with underscores between words (snake_case). For example: `first_name`, `total_price`.
- Choose meaningful names that describe what the variable represents. `x` is acceptable in a mathematical context, but `customer_age` is better in a business application.

### Examples

```python
# Good variable names
first_name = "Alice"
age = 30
total_amount = 150.75
is_active = True

# Invalid variable names (will cause errors)
# 1st_name = "Bob"        # cannot start with a digit
# first-name = "Bob"       # hyphen is not allowed
# class = "Math"           # 'class' is a reserved keyword
```

Using descriptive names makes your code self-documenting. When you read `total_amount`, you immediately understand what it represents, unlike `ta` or `x`.

## 2. Strings: Combining and Concatenating

A string is a sequence of characters enclosed in quotes. You can use single quotes (`'`) or double quotes (`"`). Both are valid, but choose one and be consistent.

### Combining Strings with the `+` Operator

You can join (concatenate) strings using the `+` operator.

```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
```

Output:
```
John Doe
```

Notice the space `" "` between the two variables. Without it, the names would be joined as `JohnDoe`.

### String Concatenation in Practice

Concatenation is useful when building messages dynamically.

```python
product = "Laptop"
price = 1200
message = "The " + product + " costs $" + str(price) + "."
print(message)
```

Output:
```
The Laptop costs $1200.
```

Note that we used `str(price)` to convert the integer to a string before concatenation. We will cover this in the numbers section.

### Alternative: f-strings (Python 3.6+)

f-strings provide a cleaner way to embed variables in strings. You prefix the string with `f` and put variables inside curly braces `{}`.

```python
first_name = "Jane"
last_name = "Smith"
full_name = f"{first_name} {last_name}"
print(full_name)
```

Output:
```
Jane Smith
```

f-strings are more readable and often preferred over concatenation. However, concatenation is still common, especially in older codebases.

## 3. Adding Whitespace to Strings with Tabs or Newlines

Whitespace refers to characters that are not visible, such as spaces, tabs, and newlines. Whitespace can make text more readable and structured.

### Tabs (`\t`)

A tab character inserts horizontal space. It is often used for indentation or aligning text.

```python
print("Name:\tAlice")
print("Age:\t30")
```

Output:
```
Name:   Alice
Age:    30
```

### Newlines (`\n`)

A newline character moves the cursor to the next line.

```python
print("Line 1\nLine 2\nLine 3")
```

Output:
```
Line 1
Line 2
Line 3
```

### Combining Tabs and Newlines

You can combine both to create structured output.

```python
print("Items:\n\tApple\n\tBanana\n\tOrange")
```

Output:
```
Items:
    Apple
    Banana
    Orange
```

This is useful for displaying lists, menus, or any structured data.

## 4. Stripping Whitespace

Often, strings contain unwanted whitespace at the beginning or end. This can happen when reading user input, processing files, or dealing with data from external sources. Stripping removes these extra spaces.

Python provides three methods for stripping whitespace:
- `lstrip()`: Removes whitespace from the left (beginning).
- `rstrip()`: Removes whitespace from the right (end).
- `strip()`: Removes whitespace from both sides.

### Examples

```python
name = "  Alice  "

print(f"'{name}'")
print(f"'{name.lstrip()}'")
print(f"'{name.rstrip()}'")
print(f"'{name.strip()}'")
```

Output:
```
'  Alice  '
'Alice  '
'  Alice'
'Alice'
```

### Practical Use Cases

1. **User Input**: When asking users for their name, they might accidentally add spaces before or after. Stripping ensures consistent data.

```python
user_input = input("Enter your name: ")
clean_name = user_input.strip()
print(f"Hello, {clean_name}!")
```

2. **Data Cleaning**: When reading data from a CSV file, fields often contain leading or trailing spaces. Stripping ensures accurate comparisons and processing.

3. **Validation**: If you are checking if a string matches a specific value, stripping prevents false negatives due to accidental spaces.

```python
answer = " yes "
if answer.strip() == "yes":
    print("User agreed")
```

Stripping is a simple but essential tool for handling real-world data.

## 5. Avoiding Syntax Errors with Strings

Syntax errors are common when working with strings, especially for beginners. They occur when the code violates the language's rules. Here are the most frequent mistakes and how to avoid them.

### 1. Unclosed String Literals

If you open a quote but forget to close it, Python will raise a syntax error.

```python
# Error
message = "Hello, World
```

**Fix**: Always close the string with the same type of quote.

```python
message = "Hello, World"
```

### 2. Mixing Quote Types

If you start with double quotes, you must end with double quotes. The same applies to single quotes. However, you can use different quotes if you need to include a quote character inside the string.

```python
# Error: mixing quotes
message = "Hello, World'
```

**Fix**: Use the same type on both ends.

```python
message = "Hello, World"
```

### 3. Including Quotes Inside Strings

If your string contains a quote, you have two options: escape it with a backslash (`\`) or use the other quote type.

```python
# Using backslash to escape
message = 'She said, "Hello"'

# Using different quote types
message = "She said, 'Hello'"
```

Both are valid. The second is often cleaner.

### 4. Concatenating Strings with Numbers Without Conversion

You cannot directly concatenate a string with a number using `+` because they are different types.

```python
# Error
age = 25
message = "I am " + age + " years old."
```

**Fix**: Convert the number to a string using `str()`.

```python
age = 25
message = "I am " + str(age) + " years old."
print(message)
```

Alternatively, use an f-string, which handles type conversion automatically.

```python
message = f"I am {age} years old."
```

### 5. Using Reserved Keywords as Variable Names

Python has reserved keywords that have special meanings. You cannot use them as variable names.

```python
# Error
class = "Math"
if = 5
```

**Fix**: Choose a different name. For example, `class_name` or `course`.

```python
class_name = "Math"
```

### 6. Missing or Extra Parentheses in `print()`

In Python 3, `print()` is a function and requires parentheses. Forgetting them causes a syntax error.

```python
# Error
print "Hello"
```

**Fix**: Add parentheses.

```python
print("Hello")
```

### 7. Indentation Errors (Though Not String-Specific)

While not directly related to strings, indentation errors are very common. Python uses indentation to define code blocks. Inconsistent indentation causes a syntax error.

```python
# Error
if True:
print("Hello")
```

**Fix**: Ensure consistent indentation (typically 4 spaces).

```python
if True:
    print("Hello")
```

Understanding these common mistakes will save you time and frustration when debugging.

- Stripping removes unwanted spaces from strings, useful for cleaning input.
