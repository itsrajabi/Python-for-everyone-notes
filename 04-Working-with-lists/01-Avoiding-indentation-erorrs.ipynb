# Chapter 04: Working with Lists
## 01 - Avoiding Indentation Errors

In the previous section, we learned how to use `for` loops to iterate over lists. Now we turn to one of the most common sources of errors for beginners: **indentation**.

Unlike many other programming languages that use braces `{}` to define code blocks, Python uses **indentation** to indicate which lines belong to which block. This makes Python code clean and readable, but it also means that incorrect indentation will cause errors or unexpected behavior.

Indentation errors are very common when you are first learning Python. Understanding them will save you hours of debugging.

## 1. Forgetting to Indent

After a `for` statement (or any statement that ends with a colon), the next line must be indented. If you forget to indent, Python raises an `IndentationError`.

### Example 1: Missing Indentation

```python
names = ["Alice", "Bob", "Charlie"]

for name in names:
print(name)  # IndentationError: expected an indented block
```

**Fix**: Indent the line inside the loop.

```python
for name in names:
    print(name)  # Correct
```

### Example 2: Multiple Lines Without Indentation

```python
for i in range(3):
print(i)  # Error
print("Done")  # This line is not indented, but Python expects it
```

**Fix**: Indent all lines that belong to the loop.

```python
for i in range(3):
    print(i)
    print("Done")
```

### Example 3: Indenting Only the First Line

```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
print("Loop finished")  # This line is NOT indented, so it runs after the loop
```

This is not an error, but it changes the behavior. The last line runs once after the loop finishes.

## 2. Forgetting to Indent Additional Lines

When you have multiple lines inside a loop, all of them must be indented at the same level. If you forget to indent one line, Python will either raise an error or treat it as outside the loop.

### Example 1: Missing Indentation on the Second Line

```python
numbers = [1, 2, 3]

for num in numbers:
    print(f"Number: {num}")
print(f"Square: {num ** 2}")  # This line is NOT indented - runs after the loop
```

Output:
```
Number: 1
Number: 2
Number: 3
Square: 9
```

The last line runs only once after the loop, using the last value of `num` (which is `3`). This is probably not what the programmer intended.

**Fix**: Indent the second line to be part of the loop.

```python
for num in numbers:
    print(f"Number: {num}")
    print(f"Square: {num ** 2}")
```

### Example 2: Partially Indented Block

```python
names = ["Alice", "Bob"]

for name in names:
    print(f"Hello, {name}")
    print(f"Welcome to the course.")
print("Enjoy learning!")  # This runs after the loop
```

Output:
```
Hello, Alice
Welcome to the course.
Hello, Bob
Welcome to the course.
Enjoy learning!
```

If the last line was intended to be inside the loop, it must be indented.

### Example 3: Inconsistent Indentation (Mixed Spaces and Tabs)

Mixing spaces and tabs can cause invisible indentation errors. Python 3 does not allow mixing tabs and spaces for indentation.

```python
for i in range(3):
    print(i)    # 4 spaces
\tprint(i)    # Tab - will cause IndentationError
```

**Fix**: Use spaces consistently. Most editors can convert tabs to spaces automatically.

## 3. Indenting Unnecessarily

Indenting code that should not be indented can cause logical errors or syntax errors. Python expects consistent indentation levels.

### Example 1: Unnecessary Indentation Outside a Block

```python
name = "Alice"
    print(name)  # IndentationError: unexpected indent
```

**Fix**: Remove the unnecessary indentation.

```python
name = "Alice"
print(name)
```

### Example 2: Unnecessary Indentation Inside a Function (Not Yet Covered)

```python
def greet():
    print("Hello")
        print("World")  # IndentationError: unexpected indent
```

**Fix**: Keep consistent indentation.

```python
def greet():
    print("Hello")
    print("World")
```

### Example 3: Unnecessary Indentation in a Conditional

```python
age = 25
if age >= 18:
    print("Adult")
    print("You can vote.")  # This is inside the if block - correct
        print("You can drive.")  # This has extra indentation - error
```

**Fix**: Keep consistent indentation for all lines in the block.

```python
if age >= 18:
    print("Adult")
    print("You can vote.")
    print("You can drive.")
```

## 4. Indenting Unnecessarily After the Loop

Sometimes beginners indent code that should run after the loop. This causes the code to run inside the loop instead of after it.

### Example 1: Indented Final Message

```python
names = ["Alice", "Bob", "Charlie"]

for name in names:
    print(f"Hello, {name}")
    print("All greetings sent.")  # This runs inside the loop!
```

Output:
```
Hello, Alice
All greetings sent.
Hello, Bob
All greetings sent.
Hello, Charlie
All greetings sent.
```

The message is printed three times because it is inside the loop.

**Fix**: Unindent the final message.

```python
for name in names:
    print(f"Hello, {name}")
print("All greetings sent.")  # Runs once after the loop
```

### Example 2: Indented Summary Calculation

```python
numbers = [10, 20, 30]
total = 0

for num in numbers:
    total += num
    print(f"Total: {total}")  # This prints after each iteration
```

Output:
```
Total: 10
Total: 30
Total: 60
```

**Fix**: Unindent to print the final total once.

```python
for num in numbers:
    total += num
print(f"Total: {total}")  # Runs once after the loop
```

Output:
```
Total: 60
```

### Example 3: Indented Final Confirmation

```python
items = ["apple", "banana", "orange"]

for item in items:
    print(f"Added: {item}")
    print("Order complete!")  # This runs after each item
```

Output:
```
Added: apple
Order complete!
Added: banana
Order complete!
Added: orange
Order complete!
```

**Fix**: Unindent the confirmation message.

```python
for item in items:
    print(f"Added: {item}")
print("Order complete!")  # Runs once after the loop
```

## 5. Forgetting the Colon

The colon (`:`) at the end of a `for` statement (or `if`, `while`, `def`, etc.) is required. Forgetting it causes a syntax error.

### Example 1: Missing Colon After `for`

```python
names = ["Alice", "Bob", "Charlie"]

for name in names  # SyntaxError: invalid syntax
    print(name)
```

**Fix**: Add the colon.

```python
for name in names:
    print(name)
```

### Example 2: Missing Colon After `if`

```python
age = 20

if age >= 18  # SyntaxError: invalid syntax
    print("Adult")
```

**Fix**: Add the colon.

```python
if age >= 18:
    print("Adult")
```

### Example 3: Multiple Missing Colons

```python
for i in range(3)  # Missing colon
    if i == 2  # Missing colon
        print("Last one")
```

**Fix**: Add colons after both statements.

```python
for i in range(3):
    if i == 2:
        print("Last one")
```

## 6. Mixed Indentation Errors

Sometimes beginners combine multiple types of indentation errors in one program.

### Example: A Program with Multiple Indentation Issues

```python
names = ["Alice", "Bob", "Charlie"]

for name in names  # Missing colon
print(name)  # No indentation
  print("Hello")  # Inconsistent indentation (2 spaces instead of 4)
    print("Done")  # Indented but should be after the loop
```

**Fix**:

```python
names = ["Alice", "Bob", "Charlie"]

for name in names:
    print(name)
    print("Hello")
print("Done")  # Runs after the loop
```

Output:
```
Alice
Hello
Bob
Hello
Charlie
Hello
Done
```

## Summary

- Python uses indentation to define code blocks. Incorrect indentation causes errors or unexpected behavior.
- **Forgetting to indent**: The line after a `for` statement must be indented.
- **Forgetting to indent additional lines**: All lines inside a block must be indented at the same level.
- **Indenting unnecessarily**: Code outside a block should not be indented.
- **Indenting after the loop**: Code that should run after the loop must be unindented.
- **Forgetting the colon**: Every `for` statement (and other block statements) must end with a colon (`:`).
- Use consistent indentation (4 spaces is the standard). Avoid mixing tabs and spaces.
- When you see an `IndentationError`, check the line indicated by the error message and the lines around it.

Understanding indentation rules will save you significant debugging time and help you write cleaner, more reliable Python code.
