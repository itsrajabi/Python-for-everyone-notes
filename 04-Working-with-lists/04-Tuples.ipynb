# Chapter 04: Working with Lists
## 04 - Tuples

In the previous sections, we worked extensively with lists. Now we turn to **tuples**, a data structure that is similar to lists but with one key difference: **tuples are immutable**. This means that once a tuple is created, its contents cannot be changed.

Tuples are useful when you want to store a collection of values that should not change throughout the program. They are often used for fixed data like coordinates, configuration settings, or constant values.

Understanding when to use tuples versus lists is an important skill in Python programming.

## 1. Defining a Tuple

A tuple is defined using parentheses `()` instead of square brackets `[]`. Elements are separated by commas.

### Example 1: Creating a Tuple

```python
# A tuple of coordinates
coordinates = (10, 20)
print(coordinates)
```

Output:
```
(10, 20)
```

### Example 2: Tuple with Different Data Types

```python
person = ("Alice", 25, True)
print(person)
```

Output:
```
('Alice', 25, True)
```

### Example 3: Single-Element Tuple

A single-element tuple requires a trailing comma.

```python
single = (5,)
print(single)
print(type(single))
```

Output:
```
(5,)
<class 'tuple'>
```

Without the comma, it is just a number in parentheses.

### Example 4: Empty Tuple

```python
empty = ()
print(empty)
```

Output:
```
()
```

### Example 5: Tuple Without Parentheses (Tuple Packing)

```python
point = 10, 20, 30
print(point)
```

Output:
```
(10, 20, 30)
```

### Example 6: Accessing Elements

Like lists, tuple elements are accessed by index.

```python
colors = ("red", "green", "blue")
print(colors[0])
print(colors[-1])
```

Output:
```
red
blue
```

## 2. Looping Through All Values in a Tuple

You can loop through a tuple just like a list using a `for` loop.

### Example 1: Looping Through a Tuple of Colors

```python
colors = ("red", "green", "blue", "yellow")

for color in colors:
    print(f"Color: {color}")
```

Output:
```
Color: red
Color: green
Color: blue
Color: yellow
```

### Example 2: Looping Through a Tuple of Numbers

```python
scores = (85, 92, 78, 90)

for score in scores:
    print(f"Score: {score}")
```

Output:
```
Score: 85
Score: 92
Score: 78
Score: 90
```

### Example 3: Using `enumerate()` with a Tuple

```python
names = ("Alice", "Bob", "Charlie")

for index, name in enumerate(names):
    print(f"Index {index}: {name}")
```

Output:
```
Index 0: Alice
Index 1: Bob
Index 2: Charlie
```

## 3. Writing Over a Tuple

You cannot change individual elements of a tuple because tuples are immutable. However, you can reassign the entire tuple to a new value.

### Example 1: Attempting to Modify a Tuple (Error)

```python
coordinates = (10, 20)
coordinates[0] = 15  # TypeError: 'tuple' object does not support item assignment
```

### Example 2: Reassigning the Entire Tuple

```python
coordinates = (10, 20)
print(f"Original: {coordinates}")

coordinates = (15, 25)
print(f"New: {coordinates}")
```

Output:
```
Original: (10, 20)
New: (15, 25)
```

### Example 3: Tuple of Mutable Elements

If a tuple contains mutable elements (like a list), you can modify those elements.

```python
data = ("John", [85, 90, 88])
data[1].append(92)
print(data)
```

Output:
```
('John', [85, 90, 88, 92])
```

The tuple itself has not changed; the list inside it has been modified.

### Example 4: Reassigning a Tuple Variable

```python
dimensions = (1920, 1080)
print(f"Original: {dimensions}")

dimensions = (2560, 1440)
print(f"New: {dimensions}")
```

Output:
```
Original: (1920, 1080)
New: (2560, 1440)
```

## 4. Differences Between Tuples and Lists

Understanding the differences between tuples and lists helps you choose the right data structure for your needs.

### Key Differences

| Feature | List | Tuple |
|---------|------|-------|
| **Mutability** | Mutable (can be changed) | Immutable (cannot be changed) |
| **Syntax** | Square brackets `[]` | Parentheses `()` |
| **Memory** | More memory (extra space for growth) | Less memory (fixed size) |
| **Speed** | Slower (more overhead) | Faster (optimized) |
| **Methods** | Many methods (`append`, `remove`, etc.) | Limited methods (`count`, `index`) |
| **Use Case** | Dynamic data that changes | Fixed data that should not change |

### Performance Considerations

Tuples are faster than lists for several reasons:

1. **Memory efficiency**: Tuples use less memory because they do not need to allocate extra space for future growth.
2. **Cache efficiency**: Smaller memory footprint means better cache performance.
3. **Creation speed**: Creating a tuple is faster than creating a list.
4. **Access speed**: Accessing elements is slightly faster in tuples because there is no need for dynamic type checking.

### When to Use Tuples vs Lists

**Use tuples when:**
- The data should not change (e.g., coordinates, configuration values).
- You need faster performance and memory efficiency.
- You want to use the tuple as a dictionary key (tuples are hashable, lists are not).

**Use lists when:**
- The data needs to change (add, remove, modify elements).
- You need to perform operations like sorting, reversing, or extending.
- The size of the data is not known in advance.

### Example: Performance Comparison (Conceptual)

```python
# List creation
list_data = [1, 2, 3, 4, 5]  # More overhead

# Tuple creation
tuple_data = (1, 2, 3, 4, 5)  # Less overhead, faster

# Accessing elements
print(list_data[2])  # Slightly slower
print(tuple_data[2])  # Slightly faster
```

## 5. Examples in Context

Here are a few practical examples showing how tuples are used in real code.

### Example 1: Coordinates in a Game

```python
player_position = (150, 300)
enemy_positions = [(50, 50), (200, 100), (400, 300)]

print(f"Player at: {player_position}")
for enemy in enemy_positions:
    print(f"Enemy at: {enemy}")
```

### Example 2: RGB Color Values

```python
colors = {
    "red": (255, 0, 0),
    "green": (0, 255, 0),
    "blue": (0, 0, 255)
}

for color_name, rgb in colors.items():
    print(f"{color_name}: {rgb}")
```

### Example 3: Configuration Settings

```python
config = ("localhost", 8080, True)
print(f"Host: {config[0]}")
print(f"Port: {config[1]}")
print(f"Debug: {config[2]}")
```

### Example 4: Returning Multiple Values from a Function

```python
def get_user_info():
    return ("Alice", 25, "alice@example.com")

name, age, email = get_user_info()
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Email: {email}")
```

Output:
```
Name: Alice
Age: 25
Email: alice@example.com
```

## Summary

- **Tuples** are immutable sequences defined with parentheses `()`.
- You can loop through tuples using `for` loops.
- You cannot modify individual elements, but you can reassign the entire tuple.
- Tuples are **faster** and more memory-efficient than lists.
- Tuples are **hashable** and can be used as dictionary keys (lists cannot).
- Use tuples for fixed data that should not change; use lists for dynamic data.
- Examples include coordinates, RGB values, configuration settings, and returning multiple values from functions.
