## 05 - Styling Your Code

In the previous sections, we learned about lists, tuples, and loops. Now we turn to an important but often overlooked topic: **code style**.

Writing code that works is one thing. Writing code that is easy to read, understand, and maintain is another. Good code style makes your code more professional, reduces bugs, and helps others (and your future self) understand what you wrote.

Python has a strong emphasis on readability, and the community has established clear guidelines for writing clean code.

## 1. The Style Guide: PEP 8

**PEP 8** stands for **Python Enhancement Proposal 8**. It is the official style guide for Python code. It was written by Guido van Rossum (the creator of Python) and other core developers.

PEP 8 is not a strict set of rules that must be followed. It is a set of recommendations that help developers write code that is consistent, readable, and maintainable. Most Python projects follow PEP 8, and many code editors and linters can check your code against it.

The Zen of Python, which can be accessed by typing `import this` in a Python interpreter, also provides guiding principles:

```python
import this
```

Key principles include:
- Beautiful is better than ugly.
- Explicit is better than implicit.
- Simple is better than complex.
- Readability counts.

### Why Follow PEP 8?

1. **Consistency**: When all developers follow the same style, code looks familiar and is easier to read.
2. **Maintainability**: Well-styled code is easier to modify, debug, and extend.
3. **Collaboration**: Teams can work together more effectively when they share coding standards.
4. **Professionalism**: Following a style guide signals that you care about quality.
5. **Fewer Bugs**: Consistent formatting helps you spot errors more easily.

## 2. Indentation

PEP 8 recommends using **4 spaces** for indentation. Tabs should be avoided because they can be interpreted differently by different editors.

### Good Indentation

```python
def greet(name):
    print(f"Hello, {name}")
    print("Welcome to Python!")

for i in range(5):
    if i % 2 == 0:
        print(f"{i} is even")
```

### Bad Indentation

```python
def greet(name):
  print(f"Hello, {name}")  # 2 spaces (inconsistent)
        print("Welcome!")   # 8 spaces (inconsistent)

for i in range(5):
    if i % 2 == 0:
    print(f"{i} is even")  # Missing indentation
```

### Continuation Lines (Line Wrapping)

When a line is too long, you can wrap it. PEP 8 recommends aligning wrapped elements.

```python
# Good: Aligned with opening delimiter
total = (first_variable
         + second_variable
         + third_variable)

# Good: Indented with 4 spaces
long_function_name(
    argument_one,
    argument_two,
    argument_three)

# Bad: No alignment
total = (first_variable
+ second_variable
+ third_variable)
```

### Hanging Indents

```python
# Good: Hanging indent
result = some_function(
    arg1, arg2, arg3,
    arg4, arg5)
```

## 3. Line Length

PEP 8 recommends limiting lines to **79 characters** for code and **72 characters** for docstrings and comments.

Why? This makes code easier to read on smaller screens, when printing, or when working with multiple files side by side.

### Example: Breaking Long Lines

```python
# Bad: Too long
message = "This is a very long message that goes on and on and on and on and on and on and on and on and on and on and on and on and on and on."

# Good: Broken into multiple lines
message = (
    "This is a very long message that goes on and on "
    "and on and on and on and on and on and on and on "
    "and on and on and on."
)

# Good: Using parentheses for line continuation
total = (first_variable + second_variable + third_variable
         + fourth_variable + fifth_variable)
```

### Breaking at Operators

When breaking lines at operators, put the operator at the beginning of the new line.

```python
# Good: Operator at the beginning
total = (first_variable
         + second_variable
         + third_variable)

# Bad: Operator at the end
total = (first_variable +
         second_variable +
         third_variable)
```

## 4. Blank Lines

Blank lines improve readability by grouping related code together.

### Guidelines for Blank Lines

- Use **two blank lines** to separate top-level functions and class definitions.
- Use **one blank line** to separate method definitions inside a class.
- Use **one blank line** to separate logical sections inside a function.
- Use **blank lines sparingly** to avoid making the code too spread out.

### Example

```python
import os
import sys


def function_one():
    print("This is function one")
    # Code here


def function_two():
    print("This is function two")
    # Code here


class MyClass:
    def method_one(self):
        print("Method one")
        # Code here

    def method_two(self):
        print("Method two")
        # Code here
```

### Inside a Function

```python
def process_data(data):
    # Step 1: Validate data
    if not data:
        return None

    # Step 2: Process data
    result = data * 2

    # Step 3: Return result
    return result
```

## 5. Other Style Guidelines

### Naming Conventions

PEP 8 recommends specific naming conventions:

| Type | Convention | Example |
|------|------------|---------|
| **Variables** | snake_case (lowercase with underscores) | `user_name`, `total_price` |
| **Functions** | snake_case | `calculate_total()`, `get_user()` |
| **Classes** | PascalCase (CapitalizedWords) | `Customer`, `ShoppingCart` |
| **Constants** | UPPERCASE with underscores | `MAX_SIZE`, `DEFAULT_COLOR` |
| **Private variables** | Leading underscore | `_internal_value` |

### Comments

Use complete sentences and proper grammar. Comments should explain the "why" not the "what".

```python
# Bad: Obvious comment
x = 5  # Assign 5 to x

# Good: Explains the reason
x = 5  # Default value for the maximum retry count

# Bad: Incomplete sentence
# calculate total

# Good: Complete sentence
# Calculate the total price including tax and shipping
```

### Whitespace in Expressions

Use spaces around operators and after commas.

```python
# Good
x = 10 + 5
items = [1, 2, 3, 4]
result = function(arg1, arg2)

# Bad
x=10+5
items=[1,2,3,4]
result=function(arg1,arg2)
```

### Trailing Whitespace

Avoid trailing whitespace at the end of lines. Most editors can show and remove it automatically.

## 6. Code Review and Reading Others' Code

Code review is the process of examining code written by others (or your own code) to find bugs, improve quality, and share knowledge. Reading others' code is also one of the best ways to improve as a programmer.

### How to Read Code Effectively

1. **Start with the big picture**: Understand what the code does before diving into details. Look for comments, function names, and the overall structure.

2. **Find the entry point**: In a program, find where execution starts (like `main()` or the first line that runs).

3. **Follow the flow**: Trace how data moves through functions. Use a notebook or comments to track variables.

4. **Identify data structures**: Understand what data is being stored and how it is organized (lists, dictionaries, tuples, etc.).

5. **Look for patterns**: Recognize common patterns like loops, conditionals, and error handling.

6. **Run the code**: If possible, run the code with test inputs and observe the output.

7. **Take notes**: Write down questions and observations. This helps you remember and understand the code.

### What to Look for in Code Review

**Correctness**: Does the code do what it is supposed to do?

**Readability**: Is the code easy to understand? Are variables named well? Is it properly formatted?

**Efficiency**: Are there obvious performance issues (e.g., unnecessary loops)?

**Edge cases**: Are there cases where the code might fail (e.g., empty lists, missing data)?

**Error handling**: Are errors handled gracefully, or does the code crash?

**Consistency**: Does the code follow the same style throughout?

**Testability**: Can the code be tested easily? Are there tests?

### Code Review Etiquette

- Be respectful and constructive.
- Focus on the code, not the author.
- Explain your reasoning, not just your opinion.
- Suggest improvements, don't just criticize.
- Ask questions when something is unclear.

### A Simple Code Review Checklist

1. Does the code follow PEP 8?
2. Are variable names descriptive?
3. Are there comments where needed?
4. Is the code free of obvious bugs?
5. Are edge cases handled?
6. Is there unnecessary duplication?
7. Is the code testable?

### Example Code Review

**Before**:

```python
def calc(x,y):
    z=x+y
    return z
```

**Feedback**:
- Function name `calc` is too vague. What is it calculating?
- Variable names `x`, `y`, `z` are not descriptive.
- No docstring explaining what the function does.
- No type hints or information about what the function expects.

**After**:

```python
def sum_two_numbers(a: float, b: float) -> float:
    """Return the sum of two numbers."""
    return a + b
```

The improved version is clearer, more descriptive, and follows best practices.

## Summary

- **PEP 8** is the official Python style guide. Following it makes your code consistent and readable.
- Use **4 spaces** for indentation. Never mix tabs and spaces.
- Keep lines to **79 characters** or less. Break long lines using parentheses or backslashes.
- Use **blank lines** to separate functions, classes, and logical sections.
- Follow **naming conventions**: `snake_case` for variables and functions, `PascalCase` for classes.
- Write clear **comments** that explain the "why", not the "what".
- **Code review** helps improve code quality and share knowledge.
- When reading code, start with the big picture, follow the flow, and take notes.
- A good code review focuses on correctness, readability, efficiency, and edge cases.

Good code style is a habit that takes practice. The more you write code, the more natural it becomes. Tools like linters (e.g., `pylint`, `flake8`) can automatically check your code against PEP 8 and help you improve.
