# Chapter 01: Getting Started
## Hello World Program

This section covers writing and running your first Python program. You will learn how to create a Python file and execute it using different tools: terminal, VS Code, PyCharm, and Spyder.

The program itself is simple:

```python
print("Hello, World!")
```

This single line of code is a tradition in programming. It confirms that your environment works and that you can write and run Python code. The `print()` function displays text on the screen.

## 1. Running from the Terminal (Using a Text Editor)

You can write Python code using a terminal-based editor like Nano or Vim, then run it from the command line.

### Steps:

1. Open a terminal.
2. Create a new Python file:
   - With Nano: `nano hello.py`
   - With Vim: `vim hello.py`
3. Type the following code:
   ```python
   print("Hello, World!")
   ```
4. Save the file and exit:
   - Nano: `Ctrl + O`, then `Ctrl + X`
   - Vim: `Esc`, then `:wq`
5. Run the program:
   ```bash
   python hello.py
   ```

You should see the output: `Hello, World!`

## 2. Running in VS Code

VS Code provides an integrated terminal and a built-in Python extension that makes running code straightforward.

### Steps:

1. Open VS Code.
2. Open the folder where you want to save your project (File → Open Folder).
3. Create a new file: `hello.py` (File → New File, then save with `.py` extension).
4. Write the code:
   ```python
   print("Hello, World!")
   ```
5. Run the program:
   - Click the triangular **Run** button (▶) at the top right.
   - Or press `Ctrl + F5` (Windows/Linux) or `Cmd + F5` (macOS).
   - Or open the integrated terminal (`Ctrl + Shift + `) and run `python hello.py`.

![VS Code Hello World](images/vscode-hello.png)

## 3. Running in PyCharm

PyCharm is a full-featured IDE with built-in run configurations.

### Steps:

1. Open PyCharm and create a new project (File → New Project).
2. In the project directory, create a new Python file: `hello.py` (right-click → New → Python File).
3. Write the code:
   ```python
   print("Hello, World!")
   ```
4. Run the program:
   - Click the green **Run** button (▶) in the top right toolbar.
   - Or right-click anywhere in the editor and select **Run 'hello'**.
   - Or press `Shift + F10`.

The output will appear in a dedicated run window at the bottom.

![PyCharm Hello World](images/pycharm-hello.png)

## 4. Running in Spyder

Spyder is designed for data science and provides an interactive environment similar to MATLAB.

### Steps:

1. Open Spyder.
2. In the editor pane, create a new file (File → New File).
3. Write the code:
   ```python
   print("Hello, World!")
   ```
4. Save the file as `hello.py` (File → Save As).
5. Run the program:
   - Click the green **Run** button (▶) on the toolbar.
   - Or press `F5`.

The output will appear in the console pane, usually on the bottom right.

![Spyder Hello World](images/spyder-hello.png)

## 5. Running in Jupyter Notebook

Jupyter Notebook provides an interactive environment where you can run code cell by cell.

### Steps:

1. Start Jupyter: in the terminal, run `jupyter notebook`.
2. In the browser interface, create a new notebook (New → Python 3).
3. In the first cell, write:
   ```python
   print("Hello, World!")
   ```
4. Run the cell: click the **Run** button or press `Shift + Enter`.

The output appears directly below the cell.

![Jupyter Hello World](images/jupyter-hello.png)

Note that Jupyter is not covered in detail in this course, but it is mentioned here because many learners use it.

## 6. Summary

- The **Hello, World!** program is a simple way to verify your environment.
- You can write Python code in any text editor or IDE.
- To run a Python file, you use the `python filename.py` command in the terminal.
- Most IDEs provide a **Run** button that does this for you.
- The `print()` function outputs text to the screen.

Now that you have successfully run your first program, you are ready to move on to more interesting topics.
