# Chapter 01: Getting Started
## Setting Up Your Programming Environment

This chapter covers the tools you need to write and run Python code. You will learn about Python versions, installation, terminal-based editors, and integrated development environments (IDEs).

## 1. Python Versions

Python is under continuous development. New versions are released regularly, with each bringing improvements, new features, and bug fixes.

### Current Stable Releases (as of mid-2026)

The Python development team maintains several release series simultaneously:

- **Python 3.14.x** – The latest feature release series. Version 3.14.6 is the most recent stable release[reference:0]. Python 3.14.0 was initially released on October 7, 2025[reference:1], and the series receives regular bug fixes and security updates.

- **Python 3.13.x** – The previous feature release series. Version 3.13.14 is the latest in this series[reference:2]. Python 3.13.0 was released on October 7, 2024[reference:3].

- **Python 3.12.x** – An older but still actively supported series. Version 3.12.13 is the latest[reference:4]. Python 3.12.0 was released on October 2, 2023[reference:5] and remains in security-fix-only mode[reference:6].

### Python 3.15 (Development)

The next major feature release, Python 3.15.0, is under active development. Pre-release versions (alpha and beta) are available for testing[reference:7]. The final release is scheduled for October 2026[reference:8].

### End-of-Life (EOL) Information

- Python 3.9 reached its end-of-life on October 31, 2025[reference:9] and is no longer supported.
- Python 3.10 and 3.11 are also approaching end-of-life and are not recommended for new projects.

### Recommendation for Beginners

For new learners, install the **latest stable release** (currently Python 3.14.x). This ensures you have access to the newest features, improved performance, and the most up-to-date documentation. All code examples in this course are compatible with Python 3.12 and above.

## 2. Installing Python

The official Python installer is available at **python.org/downloads**[reference:10]. The website automatically detects your operating system and suggests the appropriate installer[reference:11].

### Windows

1. Go to **python.org/downloads**[reference:12].
2. Click the yellow **"Download Python"** button[reference:13].
3. Run the downloaded `.exe` installer.
4. **Important**: Check the box labeled **"Add Python to PATH"** before clicking Install[reference:14][reference:15]. This allows you to run Python from the command line.
5. Follow the remaining installation steps.

### macOS

1. Go to **python.org/downloads** and download the macOS installer[reference:16].
2. Run the installer and follow the on-screen instructions[reference:17].
3. Alternatively, if you have Homebrew installed, you can run `brew install python` in the terminal[reference:18].

### Linux

Python is often pre-installed on many Linux distributions[reference:19]. However, it may be an older version. To install the latest version:

- **Ubuntu/Debian**: `sudo apt update && sudo apt install python3`
- **Fedora**: `sudo dnf install python3`
- **Arch**: `sudo pacman -S python`

You can also download the official installer from python.org for any Linux distribution.

### Verifying the Installation

Open a terminal (Command Prompt on Windows) and run:

```bash
python --version
```

This should display the installed version, for example: `Python 3.14.6`.

## 3. Terminal-Based Editors: Vim and Nano

You do not need a graphical interface to write Python code. Terminal-based text editors allow you to create and edit files directly from the command line.

### Nano

Nano is a simple, user-friendly terminal text editor[reference:20]. It is ideal for beginners who need to make quick edits from the terminal.

- To open a file: `nano filename.py`[reference:21]
- To save: `Ctrl + O`[reference:22]
- To exit: `Ctrl + X`[reference:23]

Nano displays helpful shortcuts at the bottom of the screen, making it easy to learn.

### Vim

Vim is a powerful and highly configurable text editor[reference:24]. It has a steeper learning curve but offers exceptional efficiency once mastered[reference:25].

- To open a file: `vim filename.py`[reference:26]
- To enter insert mode (to start typing): press `i`[reference:27]
- To save and exit: press `Esc`, then type `:wq` and press Enter[reference:28]

Both editors support syntax highlighting for Python, which can be enabled through configuration[reference:29].

### Why Use Terminal Editors?

Terminal editors are useful when:

- Working on remote servers via SSH
- Making quick edits without opening a full IDE
- Learning the fundamentals of the command line
- Working in environments without a graphical interface

Later in this course, you will write a Python program using a terminal editor and run it from the command line.

## 4. Integrated Development Environments (IDEs)

For larger projects and a more comfortable coding experience, most developers use an Integrated Development Environment (IDE). IDEs provide features like code completion, debugging, syntax highlighting, and project management.

### Visual Studio Code (VS Code)

VS Code is a free, open-source editor developed by Microsoft[reference:30]. It is lightweight, cross-platform, and supports many programming languages through extensions[reference:31].

- **Strengths**: Fast startup, extensive extension ecosystem, free, works with any language[reference:32]
- **Python Support**: Install the official Python extension from the marketplace
- **Popularity**: VS Code is widely used by Python developers across all experience levels[reference:33]

**Download**: code.visualstudio.com

### PyCharm

PyCharm is a professional IDE developed by JetBrains, the company behind many popular developer tools[reference:34]. It is specifically designed for Python development.

- **Strengths**: Powerful Python-specific features, excellent debugging, built-in support for testing and version control[reference:35]
- **Versions**: PyCharm Community (free) and PyCharm Professional (paid)
- **Recommendation**: PyCharm is often recommended for large Python projects due to its deep integration with the language[reference:36]

**Download**: jetbrains.com/pycharm

### Spyder

Spyder is an open-source IDE designed specifically for data science and scientific computing[reference:37]. It is included with the Anaconda distribution.

- **Strengths**: MATLAB-like interface, integrated variable explorer, works well with scientific libraries (NumPy, pandas, matplotlib)[reference:38]
- **Use Case**: Popular among data analysts, researchers, and scientists[reference:39]

**Download**: spyder-ide.org

### Which One Should You Choose?

For beginners, both VS Code and PyCharm Community are excellent choices:

- Choose **VS Code** if you want a lightweight, flexible editor that works with many languages[reference:40].
- Choose **PyCharm** if you want a dedicated Python environment with all the tools built in[reference:41].
- Choose **Spyder** if your work is primarily in data analysis or scientific computing[reference:42].

This course does not require any specific IDE. All code examples can be run from the terminal or any editor of your choice.
