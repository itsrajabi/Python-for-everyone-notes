## 02 - Comments

In the previous sections, we learned about variables, strings, and numbers. Now we turn to a practice that makes code more understandable and maintainable: writing comments.

Comments are notes you write inside your code that Python ignores. They are meant for humans — other developers, your future self, or anyone reading the code. Good comments explain the "why" behind the code, not just the "what".

## 1. How to Write Comments in Python

In Python, a comment starts with the hash symbol (`#`). Everything after `#` on that line is ignored by the Python interpreter.

```python
# This is a comment
print("Hello, World!")  # This is an inline comment
```

### Single-Line Comments

Most comments in Python are single-line comments. They can be placed on their own line or at the end of a line of code.

```python
# Calculate the total price including tax
total = price * (1 + tax_rate)
```

### Multi-Line Comments

Python does not have a dedicated multi-line comment syntax. However, you can use multiple single-line comments, or you can use a multi-line string (triple quotes) that is not assigned to a variable. The latter is often used for docstrings (documentation strings) inside functions and classes.

```python
# This is a multi-line comment
# using several single-line comments
# to explain a complex block of code.

"""
This is a docstring.
It is used to document modules, functions, and classes.
Python does not ignore it, but it is stored as metadata.
"""
```

For most comments, single-line comments are sufficient and preferred.

## 2. What Kind of Comments Should You Write?

Not all comments are useful. Some comments are redundant and add noise, while others provide valuable insight. The goal is to write comments that help the reader understand the code more quickly.

### Useful Comments

1. **Explain the "Why"**

   The code shows what is happening. Comments should explain why it is happening. This is especially important for non-obvious decisions.

   ```python
   # Use a dictionary instead of a list for faster lookups
   user_map = {user.id: user for user in users}
   ```

2. **Summarize Complex Logic**

   If a block of code is long or complicated, a brief summary helps the reader.

   ```python
   # Calculate the weighted average of sales across regions
   # Each region has a different weight based on market size
   weighted_avg = sum(sales[i] * weights[i] for i in range(len(sales)))
   ```

3. **Clarify Assumptions or Limitations**

   If your code relies on certain conditions, document them.

   ```python
   # This function assumes the input is a positive integer
   # It will not work correctly for negative values
   ```

4. **Provide Examples or References**

   If you are implementing an algorithm or using a complex formula, include a reference.

   ```python
   # Using the Black-Scholes formula for option pricing
   # See: https://en.wikipedia.org/wiki/Black–Scholes_model
   ```

### Comments to Avoid

1. **Obvious Comments**

   Do not state what the code clearly shows.

   ```python
   # Assign the value 5 to x
   x = 5  # Redundant
   ```

2. **Outdated Comments**

   Comments that are no longer accurate are worse than no comments. Always update comments when you change code.

3. **Long, Rambling Explanations**

   Be concise. If you need a long explanation, consider whether the code itself could be clearer.

4. **Comments as a Substitute for Clear Code**

   Well-written code with descriptive variable and function names often needs fewer comments. Use comments to add context, not to fix bad naming.

## 3. Comments and Clean Code

Writing good comments is part of writing clean code. Clean code is code that is easy to read, understand, and maintain. Comments contribute to clarity, but they are not a substitute for clear code.

### The Principles of Clean Code

- **Meaningful Names**: Use variable and function names that explain themselves.
- **Single Responsibility**: Each function or block should do one thing well.
- **Consistent Formatting**: Follow conventions for indentation, spacing, and naming.
- **Comments as Documentation**: Use comments to explain intent, not implementation.

### Example: Before and After

Before (poorly written):

```python
# function to calculate something
def calc(a, b):
    # add a and b
    c = a + b
    # multiply by 2
    return c * 2
```

After (clean code with meaningful comments):

```python
def double_sum(a, b):
    """Return twice the sum of a and b."""
    return (a + b) * 2
```

In the second version, the function name and docstring make the comments redundant. The code is self-documenting, and the docstring provides a clear purpose.

## 4. The Impact of Comments on the Coding Vibe and Future

Comments are not just about today — they are about the future. When you write comments, you are communicating with the developers who will work on this code after you. This might be a teammate, an open-source contributor, or your future self.

### The Vibe of Code with Good Comments

When a project has clear, thoughtful comments, it creates a welcoming atmosphere. New contributors can understand the code faster, and experienced developers can navigate the project more easily. Good comments signal that the author cares about quality and collaboration.

### The Cost of Bad or Missing Comments

Without good comments:
- Bugs are harder to find and fix.
- Onboarding new team members takes longer.
- Code becomes fragile because no one understands why certain decisions were made.
- Refactoring becomes risky.

### Career Impact

Writing clean, well-commented code is a sign of professionalism. It demonstrates that you think beyond the immediate task and consider the long-term maintainability of your work. This skill is valued in any software engineering role and is often a differentiator when working in teams or contributing to open-source projects.

In short, comments are an investment in your code's future and your reputation as a developer.

## Summary

- Comments are written with `#` and are ignored by Python.
- Good comments explain the "why" and provide context.
- Avoid obvious, outdated, or verbose comments.
- Comments are part of clean code, along with meaningful names and clear structure.
- Well-commented code is easier to maintain, collaborate on, and understand.
