# Chapter 01: Getting Started
## Running Python on Terminal

This section explains how to use Python directly from the terminal or command prompt. You will learn about the interactive Python shell, how to run single lines of code, and how to execute complete Python files from the command line.

The terminal is one of the most basic and powerful ways to interact with Python. Understanding how to use it gives you more control and helps you debug programs more effectively.

## 1. The Python Interactive Shell

When you type `python` in the terminal and press Enter, you enter the Python interactive shell (also called the REPL - Read-Eval-Print Loop).

```bash
python
```

You will see something like:

```
Python 3.14.6 (main, Mar 10 2026, 12:00:00) [MSC v.1934 64 bit (AMD64)] on win32
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

The `>>>` prompt indicates that Python is waiting for your input. You can type Python code directly and see the result immediately.

### Example: Single Line Execution

```python
>>> print("Hello, World!")
Hello, World!
>>> 5 + 3
8
>>> name = "Alice"
>>> print(f"Hello, {name}")
Hello, Alice
```

This is useful for:
- Testing small pieces of code
- Experimenting with new libraries
- Checking syntax or logic quickly
- Learning and exploring Python interactively

### Exiting the Interactive Shell

To leave the Python shell and return to the terminal, type:

```python
>>> exit()
```

Or press `Ctrl + Z` (Windows) or `Ctrl + D` (macOS/Linux).

You can also use:

```python
>>> quit()
```

Both `exit()` and `quit()` work the same way.

## 2. Running Python Files from the Terminal

While the interactive shell is useful for testing, most programs are written in `.py` files and executed as a whole.

To run a Python file from the terminal:

```bash
python filename.py
```

For example, if you have a file called `hello.py` with the following content:

```python
print("Hello, World!")
```

You would run it like this:

```bash
python hello.py
```

The output will appear directly in the terminal:

```
Hello, World!
```

This is the standard way to run Python programs. You learned this in the previous section on writing and running your first program.

## 3. Running Python on Windows

On Windows, Python is accessed through the Command Prompt or PowerShell.

### Opening the Terminal

- **Command Prompt**: Press `Win + R`, type `cmd`, and press Enter.
- **PowerShell**: Press `Win + X`, then select "Windows PowerShell" or "Terminal".

### The `python` Command on Windows

When you install Python from the official installer, you must check the option **"Add Python to PATH"**. This makes the `python` command available in any terminal window.

Without this option, you would need to type the full path to Python or use the `py` command, which is a Python launcher that comes with recent Windows installers.

### Checking Python Installation

Open Command Prompt and type:

```bash
python --version
```

This should display the installed version, for example: `Python 3.14.6`.

If you see an error, Python is not in your PATH. You can either:
- Reinstall Python and check "Add Python to PATH"
- Use the `py` command instead: `py --version`
- Add Python manually to the system PATH

### Running Python Interactively on Windows

1. Open Command Prompt or PowerShell.
2. Type `python` and press Enter.
3. You will see the `>>>` prompt.
4. Type your code and see the result immediately.
5. Type `exit()` to leave.

### Running Python Files on Windows

Navigate to the folder containing your Python file using `cd` (change directory):

```bash
cd Desktop\my_python_projects
python hello.py
```

The output will appear in the terminal.

## 4. Running Python on Mac and Linux

On macOS and Linux, the terminal is accessed through the Terminal application. Both operating systems have a Unix-like command line, so the commands are very similar.

### Opening the Terminal

- **macOS**: Press `Cmd + Space`, type `Terminal`, and press Enter.
- **Linux**: Usually found in the applications menu under "System Tools" or by pressing `Ctrl + Alt + T` on most distributions.

### The `python3` Command

On many macOS and Linux systems, both `python` and `python3` are available. The `python` command may point to an older Python 2 version on some older systems. It is recommended to use:

```bash
python3 --version
```

However, on newer systems and when using the official installer from python.org, the `python` command is set to Python 3.

### Running Python Interactively on Mac/Linux

1. Open Terminal.
2. Type `python3` and press Enter.
3. You will see the `>>>` prompt.
4. Type your code and see the result immediately.
5. Type `exit()` to leave.

### Running Python Files on Mac/Linux

Navigate to the folder containing your Python file using `cd`:

```bash
cd Desktop/python_projects
python3 hello.py
```

The output will appear in the terminal.

### Alternative: Using `python` Instead of `python3`

If you installed Python from python.org or your system defaults to Python 3, you can use `python` instead of `python3`. Check which version is installed:

```bash
python --version
```

If it shows Python 3.x, you can use `python` directly. If it shows Python 2.x, use `python3` instead.

## 5. Summary

- The **interactive shell** (REPL) is started by typing `python` (or `python3`) in the terminal.
- You can execute single lines of code directly in the interactive shell.
- To exit the interactive shell, type `exit()` or press `Ctrl + Z` (Windows) or `Ctrl + D` (macOS/Linux).
- Python files are executed by running `python filename.py` or `python3 filename.py`.
- On Windows, the `python` command works if you added Python to PATH during installation.
- On macOS and Linux, `python3` is often used instead of `python`.
- The previous section covered writing a Python file and running it from the terminal.

With these skills, you can now:
- Test small pieces of code interactively
- Write and run complete programs from the terminal
- Work with Python on any operating system

You are now ready to move on to more advanced topics.
